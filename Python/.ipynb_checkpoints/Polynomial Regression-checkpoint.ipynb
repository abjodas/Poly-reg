```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("Position_Salaries.csv")
df.tail()
```

```python
plt.figure(figsize=(12,6))
plt.scatter(x="Level", y="Salary", data=df)
plt.show()
```

# Training the Linear Regression Model

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
X = df.iloc[:, 1:2].values
y = df.iloc[:, -1].values
X
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test), 1)), 1)
```

## Polynomial Regression Model

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(2)
newX = poly.fit_transform(X)
newX
```

```python
lin_reg2 = LinearRegression()
lin_reg2.fit(newX, y)
```

### Visualizing the Linear Regression Model Results

```python
y_pred = regressor.predict(X)
plt.figure(figsize=(12,6))
plt.scatter(X, y)
plt.plot(X, y_pred)
```

### Visualizing the polynomial regression model

```python
y_pred = lin_reg2.predict(newX)
plt.figure(figsize=(12,6))
plt.scatter(X,y, color="red")
plt.plot(X, y_pred)
```

#### Polynomial Regression with degree 3

```python
poly3 = PolynomialFeatures(3)
X_poly3 = poly3.fit_transform(X)

lin_reg3 = LinearRegression()

lin_reg3.fit(X_poly3, y)
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(X, y, color="red")
plt.plot(X, lin_reg3.predict(poly3.fit_transform(X)))
```

```python
poly4 = PolynomialFeatures(4)
X_poly4 = poly4.fit_transform(X)

lin_reg4 = LinearRegression()

lin_reg4.fit(X_poly4, y)
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(X, y, color="red")
plt.plot(X, lin_reg4.predict(poly4.fit_transform(X)))
```
